Implementa el algoritmo de Salamin y Brent y compara su eficiencia con la del metodo basado en la serie de Ramanujan

```python

def phi(precision, iterations):
    NR=RealField(prec=precision)
    a, b, s=1, NR(sqrt(2)/2), 1/2
    print a, b, s
    for i in srange(iterations):
        a, b=NR((a+b)/2), NR(sqrt(a*b))
        c=NR(a**2-b**2)
        s=NR(s-(2**i)*c)
        p=NR((2*(a**2))/s)
        print p
    return p
        
```

```python
phi(100, 2000)
```

```python
def ramanujan(N, precision):
    NR=RealField(prec=precision)
    return 1/NR(sum(((factorial(2*i))^3*(42*i+5))/((factorial(i)**6)*16**(3*i+1)) for i in srange(0, N)))
```
