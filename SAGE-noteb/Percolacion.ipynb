Percolacion

```python
def moneda(prob):
    n = random()
    if(n<=prob):
        return 1
    else:
        return 0
    
def percolar(prob, NIt):
    nodosVisitados = [(0,0)]
    lAux = []
    nodosAlcanzados = [(0,0)]
    for int in xrange(NIt):
        for nodo in nodosVisitados:
            if(moneda(prob)):
                auxNode = (nodo[0], nodo[1]+1)
                lAux.append(auxNode)
            if(moneda(prob)):
                auxNode = (nodo[0], nodo[1]-1)
                lAux.append(auxNode)
            if(moneda(prob)):
                auxNode = (nodo[0]+1, nodo[1])
                lAux.append(auxNode)
            if(moneda(prob)):
                auxNode = (nodo[0]-1, nodo[1])
                lAux.append(auxNode)
        nodosAlcanzados = lAux
        for nodo in nodosAlcanzados:
            nodosVisitados.append(nodo)
        lAux=[]
    return list(set(nodosVisitados))
```
