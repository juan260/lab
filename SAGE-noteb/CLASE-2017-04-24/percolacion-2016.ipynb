```python
visitados = [(0,0)]
def prox((n,m)):
    L=[(n-1,m),(n,m-1),(n+1,m),(n,m+1)]
    L1 = []
    for P in L:
        if P not in visitados:
            L1.append(P)
    return L1
```

```python
prox((-6,11))
```

```python
def propagador(alcanzados,visitados,p):
        if alcanzados == []:
            return visitados
        else:
            alcanzados2 = []
            for P in alcanzados:
                visitados.append(P)
                L = prox(P)
                for Q in L:
                    if Q in visitados:
                        break
                    else:
                        a = random()
                        if a<p:
                            alcanzados2.append(Q)
        return propagador(alcanzados2,visitados,p)
```

```python
alcanzadosf = propagador([(0,0)],[],0.3)
```

```python
def propagador2(alcanzados,visitados,N,p):
        if alcanzados == []:
            return visitados
        else:
            alcanzados2 = []
            for P in alcanzados:
                visitados.append(P)
                L = prox(P)
                for Q in L:
                    if Q in visitados:
                        continue
                    elif not(-N<=Q[0]<=N and -N<=Q[1]<=N):
                        continue
                    else:
                        a = random()
                        if a<p:
                                alcanzados2.append(Q)
        return propagador2(alcanzados2,visitados,N,p)
```

```python
visitados1 = propagador2([(0,0)],[],10,0.45)
M1 = matriz(10,visitados1);
matrix_plot(M1)
```
